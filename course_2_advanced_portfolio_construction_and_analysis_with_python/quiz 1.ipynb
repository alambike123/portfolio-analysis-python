```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

import sys
sys.path.append("../")
import edhec_risk_kit as erk

# using seaborn style (type plt.style.available to see available styles)
plt.style.use("seaborn-dark")
```

```python
ind_rets = erk.get_ind_file(filetype="rets", nind=49)["1991":"2018"]
```

```python
ind_rets.head()
```

```python
rets = pd.read_csv("../data/F-F_Research_Data_Factors_m.csv",
                   header=0, index_col=0, na_values=-99.99).loc["1991":"2019"]/100
rets.index = pd.to_datetime(rets.index, format="%Y%m").to_period('M')
```

```python
rets['Constant'] = 1
x_var = rets.loc[:, ['Mkt-RF', 'Constant']]
x_var.head()
```

```python
beer = ind_rets['Beer'] - rets['RF']
```

```python
# 1
lm = sm.OLS(beer, x_var).fit()
lm.summary()
```

```python
# 2
steel= ind_rets['Steel'] - rets['RF']
lm = sm.OLS(steel, x_var).fit()
lm.summary()
```

```python
# 3
x_var = rets.loc["2013":"2018", ['Mkt-RF', 'Constant']]
x_var.head()

beer = ind_rets.loc["2013":"2018",'Beer'] - rets.loc["2013":"2018",'RF']

lm = sm.OLS(beer, x_var).fit()
lm.summary()
```

```python
# 4
steel = ind_rets.loc["2013":"2018",'Steel'] - rets.loc["2013":"2018",'RF']

lm = sm.OLS(steel, x_var).fit()
lm.summary()
```

```python
# 5, 6
y_vars = ind_rets.loc['1991': '1993':]
x_var = rets.loc['1991': '1993', ['Mkt-RF', 'Constant']]


y_vars2 = y_vars - rets.loc['1991':'1993', 'RF'].values.reshape((36,1))

betas = []
ind_names = []
for col in y_vars2.columns:
    fit = sm.OLS(y_vars2[col], x_var).fit()
    
    betas.append(fit.params[0])
    ind_names.append(col)

ind_betas = pd.DataFrame({'ind': ind_names, 'beta': betas})
```

```python
ind_betas.sort_values(by='beta')
```

```python
# 7, 8
y_vars = ind_rets.loc['1991': '2018',]
x_var = rets.loc['1991':'2018',]


y_vars2 = y_vars - rets.loc['1991':'2018', 'RF'].values.reshape((336,1))

coeff_2 = []
ind_names = []
for col in y_vars2.columns:
    fit = sm.OLS(y_vars2[col], x_var).fit()
    
    coeff_2.append(fit.params[1])
    ind_names.append(col)

coeff_2_df = pd.DataFrame({'ind': ind_names, 'coeff_smb': coeff_2})
```

```python
coeff_2_df.sort_values(by='coeff_smb')
```

```python
# 9, 10
y_vars = ind_rets.loc['1991': '2018',]
x_var = rets.loc['1991':'2018',]


y_vars2 = y_vars - rets.loc['1991':'2018', 'RF'].values.reshape((336,1))

coeff_2 = []
ind_names = []
for col in y_vars2.columns:
    fit = sm.OLS(y_vars2[col], x_var).fit()
    
    coeff_2.append(fit.params[2])
    ind_names.append(col)

coeff_2_df = pd.DataFrame({'ind': ind_names, 'coeff_vmg': coeff_2})
display(coeff_2_df.sort_values(by='coeff_vmg'))
```
